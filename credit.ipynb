Many features have skewed distributions, we have to mitigate this skewness.
Why skewness has to be mitigated:
1. PCA doesn’t fix class imbalance – it only transforms features, not target distribution.
2. Prevents model bias – avoids models predicting only the majority (non-fraud) class.
3. Improves learning and evaluation – helps capture fraud patterns and gives fairer metrics (F1, AUC).
4. Ensures stable model performance – reduces training bias and improves generalization to new data.

```python
# Removing skewness using Power Transformer
from sklearn.preprocessing import PowerTransformer

# Initialize PowerTransformer with Yeo-Johnson method
power_transformer = PowerTransformer(method='yeo-johnson', standardize=True, copy=False)

# Apply transformation to training and testing data
X_train[X_train.columns] = power_transformer.fit_transform(X_train[X_train.columns])
X_test[X_test.columns] = power_transformer.transform(X_test[X_test.columns])
```

Before model training, we first convert our imbalanced dataset into a balanced dataset.

Why dataset needs to be balanced before model training:
1. An imbalanced dataset biases the model toward the majority (non-fraud) class, leading to poor detection of minority (fraud) cases.
2. Balancing the data ensures sufficient representation of both classes, allowing the model to learn meaningful fraud patterns.
3. It provides a more accurate evaluation of model performance through metrics like F1-score and PR-AUC instead of misleading accuracy.
4. Balanced data helps the model establish clearer decision boundaries and reduces overfitting to dominant patterns

Techniques to balance dataset:
1. Oversampling
2. Undersampling
3. SMOTE (Synthetic minority oversampling technique)

**Why we don't use SVM and KNN here?**

Support Vector Machines (SVM) are generally not preferred when working with large datasets because the algorithm requires significant computational power to perform the optimization and construct the separating hyperplane. The complexity increases further when using kernel transformations. Additionally, applying K-Fold cross-validation for hyperparameter tuning with SVM considerably increases the training time, making it impractical when computational resources are limited. For these reasons, SVM was not used in this case.

K-Nearest Neighbors (KNN) is also inefficient for large datasets because it is a lazy learner, meaning it does not build a model during training and instead stores the entire dataset in memory. During prediction, the algorithm must compute the distance between the new instance and every stored data point, which becomes computationally expensive as the dataset grows. This leads to slow predictions and high memory usage, making KNN unsuitable for large-scale applications.

```python
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    f1_score, accuracy_score, RocCurveDisplay
)
import matplotlib.pyplot as plt
import pandas as pd

def train_model(model, X_train, y_train):
    """Fit any sklearn model."""
    model.fit(X_train, y_train)
    return model


def evaluate_model(model, X_train, y_train, X_test, y_test):
    """Evaluate on both train and test sets."""
    
    # Train predictions
    y_train_pred = model.predict(X_train)
    # Test predictions
    y_test_pred = model.predict(X_test)

    print("\n=== TRAIN PERFORMANCE ===")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
    print(f"Train ROC-AUC: {roc_auc_score(y_train, model.predict_proba(X_train)[:,1]):.4f}")
    print(f"Train F1 Score: {f1_score(y_train, y_train_pred):.4f}")
    print(f"Train Accuracy: {accuracy_score(y_train, y_train_pred):.4f}")

    print("\n=== TEST PERFORMANCE ===")
    print(confusion_matrix(y_test, y_test_pred))
    print(classification_report(y_test, y_test_pred))
    print(f"Test ROC-AUC: {roc_auc_score(y_test, model.predict_proba(X_test)[:,1]):.4f}")
    print(f"Test F1 Score: {f1_score(y_test, y_test_pred):.4f}")
    print(f"Test Accuracy: {accuracy_score(y_test, y_test_pred):.4f}")

    # ROC Curve
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title("ROC Curve")
    plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, KFold

def tune_model(model, param_grid, X_train, y_train, scoring='roc_auc'):
    """Perform grid search with KFold."""
    
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        scoring=scoring,
        cv=kf,
        n_jobs=-1,
        verbose=1,
        return_train_score=True
    )
    
    grid_search.fit(X_train, y_train)

    print("Best Parameters:", grid_search.best_params_)
    
    # Visualize tuning behavior
    cv_results = pd.DataFrame(grid_search.cv_results_)
    plt.figure(figsize=(8,5))
    plt.plot(cv_results['param_' + list(param_grid.keys())[0]], cv_results['mean_test_score'], label="Validation Score")
    plt.plot(cv_results['param_' + list(param_grid.keys())[0]], cv_results['mean_train_score'], label="Training Score")
    plt.xlabel("Parameter Value")
    plt.ylabel("ROC-AUC Score")
    plt.xscale('log')
    plt.legend()
    plt.title("Hyperparameter Tuning Performance")
    plt.show()
    
    return grid_search.best_estimator_
```

```python
def compare_models(models, X_test, y_test):
    """
    Compare multiple models on Accuracy, F1 Score, and ROC-AUC.
    
    models: dict -> {'Model Name': trained_model, ...}
    """

    results = []

    for name, model in models.items():
        y_pred = model.predict(X_test)

        # If model supports predict_proba use that; else fallback to decision_function
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_test)[:, 1]
        else:
            y_prob = model.decision_function(X_test)

        results.append([
            name,
            accuracy_score(y_test, y_pred),
            f1_score(y_test, y_pred),
            roc_auc_score(y_test, y_prob)
        ])

    df_results = pd.DataFrame(results, columns=['Model', 'Accuracy', 'F1 Score', 'ROC-AUC'])
    df_results.set_index('Model', inplace=True)

    display(df_results)

    # Plot comparison
    df_results.plot(kind='bar', figsize=(10,6))
    plt.title("Model Performance Comparison")
    plt.ylabel("Score")
    plt.ylim(0, 1)
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.show()

    return df_results
```

```python
import os
import joblib

def save_model(model, folder_path, file_name):
    """
    Save model to a specific folder.
    If the folder does not exist, it will be created.
    """
    # Create folder if it doesn't exist
    os.makedirs(folder_path, exist_ok=True)
    
    file_path = os.path.join(folder_path, file_name)
    joblib.dump(model, file_path)
    
    print(f"Model saved successfully at: {file_path}")
```

```python
def evaluate_model_with_confusion_metrics(model, X_test, y_true):
    """
    Takes a trained model, X_test, and y_true.
    Computes y_pred, confusion matrix, and prints:
    - TPR (Recall / Sensitivity)
    - FPR
    - TNR (Specificity)
    - Precision
    """
    # Predict using the model
    y_pred = model.predict(X_test)

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    # Metrics
    tpr = tp / (tp + fn) if (tp + fn) != 0 else 0   # Sensitivity / Recall
    fpr = fp / (fp + tn) if (fp + tn) != 0 else 0
    tnr = tn / (tn + fp) if (tn + fp) != 0 else 0   # Specificity
    precision = tp / (tp + fp) if (tp + fp) != 0 else 0

    # Print results
    print("Model Evaluation Metrics: ")
    print(f"TPR (Recall):       {tpr:.3f}")
    print(f"FPR:                {fpr:.3f}")
    print(f"TNR (Specificity):  {tnr:.3f}")
    print(f"Precision:          {precision:.3f}")
```

Different balancing techniques were applied namely Oversampling, Undersampling and SMOTE and several models were trained including Logistic Regression, Decision Trees, Random Forest and XGBoost.

From the comparison tables we infer:

| Technique     | Best Model            | Accuracy | F1 Score   | ROC-AUC    |

**Analysis**

1. Undersampling leads to loss of data and information as it reduces the majority class. Though the models performed decently, the F1-scores are much lower than Oversampling or SMOTE so it's not ideal.

2. Oversampling and SMOTE both produced excellent ROC-AUC scores showing that the models can distinguish between the 2 classes very well.

3. Among all the models trained and evaluated so far, XGBoost achieved the highest F1-scores (0.85 - 0.89) and very high ROC-AUC ( > 0.98) indicating excellent precision - recall balance and strong discrimination ability.
